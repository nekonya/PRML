# 前書き

* パラメトリック(parametric)なアプローチ
  - 確率分布を少数のパラメータで形状を表せる密度関数で表現
    - 選択した密度関数の表現力が不足する場合には，予測性能が悪くなり得る
      - 例）多峰性分布から生成されたデータ → 単峰性のガウス分布で捉えることには無理がある
* ノンパラメトリック(nonparametric)なアプローチ
  - 分布形状について少数の仮定のみ
  - 以降では簡単な頻度主義的な手法を扱う
    - 近年ではベイズ的な手法が注目されてきている

# ヒストグラム密度推定法

* ヒストグラム密度推定法(histogram density estimation method)
  - 既に図1.11や図2.6で扱われている

* 以降では，１つの連続変数$x$のみのデータについて議論
  - $x$を幅$\Delta_i$の別々の区間に区切る
  - $i$番目に入った$x$の観測値の数$n_i$を数える
  - $N$は観測値の総数とすると，各区間$i$の確率分布は以下に表される
    $$p_i = \frac{n_i}{N \Delta_i}$$
  - 正規化されていることは以下に確認できる．
    $$ \int p(x) dx \to \sum_i \Delta_i p_i = \sum_i \Delta_i \frac{n_i}{N \Delta_i} = \frac{1}{N} \sum_i n_i = 1$$
  - 多くの場合は$\Delta_i = \Delta$と同じ幅になるように選択される

* 例）２つのガウス分布の混合分布
  - $\Delta$の選択によって得られる結果が変わる
    - $\Delta$が小さすぎる → まばらなトゲトゲ状，元の確率分にはない構造が生じてしまう
    - $\Delta$が大きすぎる → なだらかになりすぎて，二峰性の特徴が消失
    - 適切な$\Delta$を選択すると良い結果が得られる
  - 区間の縁の選び方でも結果が変わる

* ヒストグラム密度推定の利点
  - ヒストグラムが得られれば元のデータは要らなくなる
  - データが逐次的に得られる場合でも容易に適応できる

* 問題点
  - 区間の縁で不連続性が表れる（元のデータの性質に依らず縁の選択で形状が変わり得る）．
  - 次元が増大すると必要となるデータ量が爆発する
    - D次元空間をM分割すると，区間の数は$M^D$ → 次元の増大に対して指数関数的に増加（次元の呪い）
    - 1〜2次元程度のデータの容易な可視化には役立つが，それ以上には適さない．

* 知見
  - 特定の位置での確率密度の推定には，その点の局所的な近傍にある他のデータ点も考慮する必要がある
  - → 近傍の性質は区間の選択で定義される
    - → 区間の幅が*平滑化パラメータ(smoothing parameter)*
  - 平滑化パラメータは適切に選択する必要がある
    - 大きすぎても小さすぎてもいけない．適切な中間的な値を選択する必要がある．

# カーネル密度推定法

あるD次元ユークリッド空間中の未知の確率密度$p(x)$から生成された観測値の集合が得られている．
→ 観測値の集合から$p(x)$を推定したい．

D次元ユークリッド空間中の（$x$を含む）小さな領域$R$を導入する．xが$R$に含まれる確率は

$$P = \int_R p(x) dx$$

観測値の総数を$N$とすると，$R$に含まれる観測値の数$K$は二項分布に従う．

$$Bin(K | N, P) = \frac{N!}{K! (N - K)!} P^K (1 - P)^{N-K}$$

二項分布の平均値

$$E[m] = \sum_{m-0}^{N} m Bin(m|N, \mu) = N \mu$$

から，xが領域$R$にある平均割合は$E[K/N] = P$となる．

分散は

$$var[m] = \sum_{m=0}^N (m-E[m])^2 Bin(m|N,\mu) = N \mu (1-\mu)$$

から，$var[K/N] = P(1-P)/N$となる．これより，$N$が大きくなるとこの分布は平均の周囲で細く鋭いものになる．
よって，$K$は以下の様に評価される．

$$K \sim NP$$

ここでp(x)がほとんど一定とみなせる程に$R$が小さいと仮定すると，

$$P \sim p(x) V$$

$V$は$R$の体積である．これより

$$p(x) = \frac{K}{N V} \ \ \ \ (*)$$

が得られる．

以上の議論が以下の相反する仮定に依存していることに注意する．
* 領域内でp(x)が一定とみなせる程に$R$は小さい．
* 二項分布が鋭く尖る程に$K$が大きくなる．

(*)は２通りに扱うことができる
* $V$を固定して，データから$K$の値を推定する
  - → カーネル密度推定
* $K$を固定して，データから$V$の値を推定する
  - → K近傍密度推定

どちらも$V$が$N$に応じて縮小し$K$が増大するなら，$N \to \infty$の極限で真の確率分布に収束する．

確率密度を求めたい位置$x$を中心とする小さな超立方体を領域$R$とする．
単位立方体内に含まれるか否かを判別する関数（カーネル関数(kernel function)，Parzen窓(Parzen window)）を用意する．

$$k(u) = \left\{ \begin{array}{l} 1, \ \ \ |u_i| \le \frac{1}{2} \\ 0, \ \ \ otherwise\end{array} \right.$$

$x$を中心とする１辺$h$の立方体内への包含を判定する関数は$k((x - x_n)/h)$と書ける．
この立方体内に含まれるデータ点数は以下に表される．

$$K = \sum_{n=1}^N k(\frac{x-x_n}{h})$$

$V=h^D$とすると，確率密度$p(x)$は

$$p(x) = \frac{1}{N} \sum_{n=1}^N \frac{1}{h^D} k(\frac{x-x_n}{h})$$

これはParzen推定法(Parzen estimator)と呼ばれる．$k(u)$の対称性から拡大解釈すると，中心が$x_n$にある$N$個の超立方体についての総和と解釈できる．

カーネル関数をなめらかなガウスカーネル（$h$は標準偏差）にすると以下の確率密度モデルになる．

$$p(x) = \frac{1}{N} \sum_{n=1}^N \frac{1}{(2\pi h^2)^{D/2}} \exp{-\frac{||x - x_n||^2}{2h^2}}$$

これは各観測点を中心とするガウシアンの影響を全ての観測点について足し正規化したものになっている．

* 例）２つのガウス分布の混合分布
  - $h$の選択によって得られる結果が変わる
    - $h$が小さすぎる → まばらなトゲトゲ状，元の確率分にはない構造が生じてしまう
    - $h$が大きすぎる → なだらかになりすぎて，二峰性の特徴が消失
    - 適切な$h$を選択すると良い結果が得られる
  - $h$の選び方でも結果が変わる → $h$の最適化はモデル複雑度に対応

![kernel](PRML/Figure2.25.png "カーネル法")

カーネル関数$k(u)$は以下の条件を満たす任意の関数が選択可能

$$k(u) \ge 0$$
$$\int k(u) du = 1$$

これよりカーネルを用いた確率密度が全定義域で非負，全定義域での積分値が1になることが保証される．

カーネル密度推定は訓練段階で訓練集合を保存するだけで計算を必要としない．しかし，密度評価計算コストは訓練データのデータ量に比例する欠点がある．

# 最近傍法

カーネル密度推定の欠点である「$h$を小さめにすると密度の低い領域でノイズの多い推定をしてしまう」を回避するために「密度に応じて$h$を設定する」．

$K$を固定して，データから$V$の値を推定する．

確率分布$p(x)$を推定したい点$x$を中心とした超小球を考える．
小球の半径は内部に観測値の数が$K$個になるように選択する．
この時の小球の体積を$V$とすると$p(x)$は以下で表される．

$$p(x) = \frac{K}{N V}$$

![k-neighbour](PRML/Figure2.26.png "K近傍法")

## K近傍法のクラス分類問題への拡張

クラス$C_k$に$N_k$個の観測値がある．$\sum_k N_k = N$．

ここで新たな点$x$を分類する．$x$を中心とした$K$個の観測値を含む小球を見つける．
各クラスの密度推定値は

$$p(x|C_k) = \frac{K_k}{N_k V}$$．

クラス条件のない確率密度は

$$p(x) = \frac{K}{N V}$$．

クラスの事前分布は

$$p(C_k) = \frac{N_k}{N}$$

ベイズの定理を適用すると，クラスに帰属する事後確率が得られる．

$$p(C_k|x) = \frac{p(x|C_k) p(C_k)}{p(x)} = \frac{K_k}{K}$$

誤分類率を最小にするには事後確率を最大にするクラスを選択する．

![explanation of k-neighbour part 1](PRML/Figure2.27a.png "K近傍法の説明 1")
![explanation of k-neighbour part 2](PRML/Figure2.27b.png "K近傍法の説明 2")

- $K$の選択によって得られる結果が変わる
  - $K$が小さすぎる → 細かな構造が表れる
  - $K$が大きすぎる → 構造が大雑把になりすぎる
- $K=1$でも$N \to \infty$で誤分類率が達成可能な最小誤分類率のたかだか２倍にしかならない．
  - 真の確率分布から推定する分類器を「達成可能な最小誤分類識別器」
